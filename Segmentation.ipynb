```python
import pandas as pd
import numpy as np
import cv2
from glob import glob
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F

from torch.utils.data import Dataset, DataLoader

from torchvision.io.image import read_image
from torchvision import models
from torchvision.models.segmentation import fcn_resnet50, FCN_ResNet50_Weights,\
                                            DeepLabV3_MobileNet_V3_Large_Weights, deeplabv3_mobilenet_v3_large,\
                                            fcn_resnet101, FCN_ResNet101_Weights
from torchvision.transforms.functional import to_pil_image
from torchvision.transforms import v2, InterpolationMode

import albumentations as A
from segmentation_models_pytorch.losses import DiceLoss
import segmentation_models_pytorch as smp

from sklearn.model_selection import train_test_split, ShuffleSplit
from sklearn.utils import shuffle

import matplotlib.pyplot as plt
```

```python
inputs_path = "./data/ISIC2018_Task1-2_Training_Input"
outputs_path = "./data/ISIC2018_Task1_Training_GroundTruth"
```

```python
#  prepare model class
model = fcn_resnet50(weights=FCN_ResNet50_Weights.DEFAULT)
model.classifier.append(nn.Conv2d(21,1,1,1))
model.aux_classifier.append(nn.Conv2d(21,1,1,1))
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
model = model.to(device)
```

```python
IMG_SIZE = 224

transform = v2.Compose([v2.Resize((224, 224))])

def get_train_augs():
    return A.Compose([
        A.Resize(IMG_SIZE, IMG_SIZE),
        A.HorizontalFlip(p=0.5),      # Horizontal Flip with 0.5 probability
        A.VerticalFlip(p=0.5)         # Vertical Flip with 0.5 probability
    ], is_check_shapes=False)

def get_val_augs():
    return A.Compose([
        A.Resize(IMG_SIZE, IMG_SIZE)
    ], is_check_shapes=False)
```

```python
class HAMImageDataset(Dataset):
    def __init__(self, train_path, labels_path, augs=None):
        self.images = [path for path in glob(train_path + "/*.jpg")]
        self.labels = [path for path in glob(labels_path + "/*.png")]
        
        if len(self.images) != len(self.labels):
            print(len(self.images), len(self.labels))
            raise ValueError
        
        self.augs = augs
            
    def __len__(self):
        return len(self.images)
    
    def __getitem__(self, idx):
        image = read_image(self.images[idx])
        if image.shape[0] == 4:
            image = image[:3, :, :]
        label = read_image(self.labels[idx]).squeeze(0)
        if self.augs:
            data = self.augs(image=image.permute(1,2,0).numpy(), mask=label.numpy())
            image = np.transpose(data['image'], (2,0,1))
            label = data['mask']
        label = label / 255.0
        
        return torch.tensor(image), torch.tensor(label).round().unsqueeze(0)
```

```python
ds = HAMImageDataset(inputs_path, outputs_path, augs=get_train_augs())
ds_train = HAMImageDataset("./data/train/inputs", "./data/train/ground_truth", augs=get_train_augs())
ds_test = HAMImageDataset("./data/test/inputs", "./data/test/ground_truth", augs=get_train_augs())
x, y = ds_train[3]
```

```python
train_dataloader = DataLoader(ds_train, batch_size=16, shuffle=True)
test_dataloader = DataLoader(ds_test, batch_size=8, shuffle=False)
```

```python
#  fidelity metrics
def accuracy(pred, mask):
    """accuracy metric for calculating common accuracy"""
    pred = pred.view(pred.size(0), -1)
    mask = mask.view(mask.size(0), -1)
    
    positives = (pred == mask).sum()
    N = pred.size(0) * pred.size(1)
    return positives / N 


def iou(pred, target):
    intersection = (pred.eq(target.long())
                    .type_as(pred)
                    .mul(target)
                    .sum()
                   )
    union = pred.sum() + target.sum() - intersection
    return intersection.float() / union


def dice_score(pred, target, smooth=1):  
    intersection = torch.sum(pred.flatten() * target.flatten(), dim=0)
    return (2. * intersection + smooth) / (torch.sum(pred + target) + smooth)


def rec_prec(pred, target):
    pred, target = pred.long(), target.long()
    tp, fp, fn, tn = smp.metrics.get_stats(pred, target, mode='binary', threshold=0.5)
    f1_score = smp.metrics.f1_score(tp, fp, fn, tn, reduction="micro")
    f2_score = smp.metrics.fbeta_score(tp, fp, fn, tn, beta=2, reduction="micro")
    precision = smp.metrics.precision(tp, fp, fn, tn, reduction="macro")
    recall = smp.metrics.recall(tp, fp, fn, tn, reduction="micro-imagewise")
    return f1_score.item(), f2_score.item(), precision.item(), recall.item()
```

```python
def train(model, loader, optimizer, criterion):
    model.train()
    total_loss = 0
    total_iou = 0
    total_dice = 0
    
    total_f1 = 0
    total_f2 = 0
    total_prec = 0
    total_rec = 0
    
    for i, batch in enumerate(loader):
        inputs, masks = batch
        inputs = inputs.float().to(device)
        masks = masks.float().to(device)
        
        outputs = model(inputs)["out"]
        
#         outputs = torch.nn.functional.softmax(outputs)
        
        flat_out = outputs
        flat_masks = masks
        
        loss = criterion(torch.nn.functional.sigmoid(flat_out).round(), flat_masks)
        
        total_loss += loss.item()
        total_dice += dice_score(torch.nn.functional.sigmoid(flat_out).round(), flat_masks).item()
        total_iou += iou(torch.nn.functional.sigmoid(flat_out).round(), flat_masks).item()
        
        f1, f2, prec, rec = rec_prec(torch.nn.functional.sigmoid(flat_out).round(), flat_masks)
        
        total_f1 += f1
        total_f2 += f2
        total_prec += prec
        total_rec += rec
        
        loss.backward()
        optimizer.step()
        
        optimizer.zero_grad()
        
        if (i + 1) % 20 == 0:
            print("-" * 90)
            print("| loss:", (total_loss / (i + 1)), "| prec:", (total_prec / (i + 1)), "recall:", (total_rec / (i + 1)), " dice:", (total_dice / (i + 1)), " | iter: [" + str((i + 1)) + "/" \
                  + str(len(loader)) + "]|")
    return total_loss / len(loader), total_dice / len(loader), total_iou / len(loader), total_f1 / len(loader), total_prec / len(loader), total_rec / len(loader)
```

```python
def evaluate(model, loader, criterion):
    model.eval()
    total_loss = 0
    total_iou = 0
    total_dice = 0
    
    total_f1 = 0
    total_f2 = 0
    total_prec = 0
    total_rec = 0
    
    with torch.no_grad():
        for i, batch in enumerate(loader):
            inputs, masks = batch
            inputs = inputs.float().to(device)
            masks = masks.float().to(device)
        
            outputs = model(inputs)["out"]
        
#         outputs = torch.nn.functional.softmax(outputs)
        
            flat_out = outputs
            flat_masks = masks
        
            loss = criterion(torch.nn.functional.sigmoid(flat_out).round(), flat_masks)
        
            total_loss += loss.item()
            total_dice += dice_score(torch.nn.functional.sigmoid(flat_out).round(), flat_masks).item()
            total_iou += iou(torch.nn.functional.sigmoid(flat_out).round(), flat_masks).item()
        
            f1, f2, prec, rec = rec_prec(torch.nn.functional.sigmoid(flat_out).round(), flat_masks)
        
            total_f1 += f1
            total_f2 += f2
            total_prec += prec
            total_rec += rec
        
            if (i + 1) % 20 == 0:
                print("-" * 90)
                print("| loss:", (total_loss / (i + 1)), "| prec:", (total_prec / (i + 1)), "recall:", (total_rec / (i + 1)), " dice:", (total_dice / (i + 1)), " | iter: [" + str((i + 1)) + "/" \
                      + str(len(loader)) + "]|")
    return total_loss / len(loader), total_dice / len(loader), total_iou / len(loader), total_f1 / len(loader), total_prec / len(loader), total_rec / len(loader)
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=1e-5)
criterion1 = nn.BCEWithLogitsLoss()
criterion2 = DiceLoss(mode='binary')
criterion = lambda o, gt: criterion1(o, gt) + criterion2(o, gt)
epochs = 10
min_loss = np.inf
for epoch in range(1, epochs + 1):
    print(f"Starting epoch {epoch} training...")
    loss, dice, IoU, f1, precision, recall = train(model, train_dataloader, optimizer, criterion)
    print("=" * 90)
    print("train loss:", loss, "train dice:", dice, "train IoU:", IoU, "train f1:", f1, "train precision:", precision, "train recall:", recall)
    test_loss, test_dice, test_IoU, test_f1, test_prec, test_rec = evaluate(model, test_dataloader, criterion)
    print("test loss:", test_loss, "test dice:", test_dice, "test IoU:", test_IoU, "test f1:", test_f1, "test precision:", test_prec, "test recall:", test_rec)
    print("+" * 90)
    if test_loss < min_loss:
        min_loss = test_loss
        print("BEST ATTEMT:", min_loss)
        torch.save(model.state_dict(), "fcn_resnet50-ham.pt")
    print("=" * 90)
```
